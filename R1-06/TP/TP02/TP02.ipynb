### Programme 2:

```python
h = float(input('Quelle hauteur ?'))
n = 0
epaisseur = 0.2
while epaisseur < h:
    epaisseur = epaisseur*2
    n = n+1
print(n)
```

Il faut 11 pliages pour dépasser la hauteur de la tour effeil
